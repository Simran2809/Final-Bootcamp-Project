# Min Max Scaler

```python
# Performing standard scaling on numerical variables
# Using min max scaler package of sklearn library to feature scale our data

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(numerical_features)
feature_scaled_variables_array = scaler.transform(numerical_features)
feature_scaled_variables_array
```

```python
# Coverting array to dataframe

df_feature_scaled_variables = pd. DataFrame(feature_scaled_variables_array, columns=numerical_features.columns)
df_feature_scaled_variables
```

```python
# Dropping original values of numerical variables as we will now use only feature scaled values
df.drop(columns=numerical_features.columns, inplace=True)
```

```python
# Concatenating the feature scaled values of the numerical features with the original dataframe

df = pd.concat([df.reset_index(drop=True), df_feature_scaled_variables], axis=1)
df
```
